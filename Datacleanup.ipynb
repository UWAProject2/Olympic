```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
from config import username,password
```

```python
#importing athlete csv file
csv_file = "resources/athlete_events.csv"
athlete_data_df = pd.read_csv(csv_file)
athlete_data_df.head()
```

```python
#display the info before dataframe cleanup
athlete_data_df.info()
```

```python
#importing noc csv file
csv_file = "resources/noc_regions.csv"
noc_data_df = pd.read_csv(csv_file)
noc_data_df.head()
```

# Clean DataFrame

```python
#remove rows that have got NA under medal columns as we only analysis athletes who won medal
athlete_data_df.dropna(subset=['Medal'],inplace=True)
```

```python
#display the dataset after remove na from Medal column
athlete_data_df.head()
```

```python
#rename column name for athlete data df
athlete_data_df.rename(columns={'ID': 'athlete_id'}, inplace=True)
athlete_data_df.rename(columns={'Name': 'name'}, inplace=True)
athlete_data_df.rename(columns={'Sex': 'sex'}, inplace=True)
athlete_data_df.rename(columns={'Age': 'age'}, inplace=True)
athlete_data_df.rename(columns={'Height': 'height'}, inplace=True)
athlete_data_df.rename(columns={'Weight': 'weight'}, inplace=True)
athlete_data_df.rename(columns={'Team': 'team'}, inplace=True)
athlete_data_df.rename(columns={'NOC': 'noc_id'}, inplace=True)
athlete_data_df.rename(columns={'Year': 'year'}, inplace=True)
athlete_data_df.rename(columns={'Season': 'season'}, inplace=True)
athlete_data_df.rename(columns={'City': 'city'}, inplace=True)
athlete_data_df.rename(columns={'Sport': 'sport'}, inplace=True)
athlete_data_df.rename(columns={'Event': 'event'}, inplace=True)
athlete_data_df.rename(columns={'Medal': 'medal'}, inplace=True)
```

```python
#display dataframe
athlete_data_df.head()
```

```python
#checked duplicated information
duplicaterows_df = athlete_data_df[athlete_data_df.duplicated()]
duplicaterows_df.head()
```

```python
#As the data in Year 1900 is duplicated for few athletes. Hence, the analysis start from year 1916 to give more accurate analysis.

athlete_data_df = athlete_data_df[athlete_data_df['year'] >= 1916]
```

```python
athlete_data_df.head()
```

```python
#Deleting columns that we are not using for our analysis - We deleted Height, Weight and Games
athlete_data_df.drop(columns=['height', 'weight', 'Games'], inplace=True)
```

```python
#display dataframe
athlete_data_df.head()
```

```python
#Dropping the notes column on NOC dataframe 
noc_data_df.drop(columns=['notes'], inplace=True)
```

```python
#diplay dataframe
noc_data_df.head()
```

```python
#rename column name for noc data df
noc_data_df.rename(columns={'NOC': 'noc_id'}, inplace=True)
```

```python
#diplay dataframe
noc_data_df.head(15)
```

```python
#Binned the data into age ranges 
age_bins = [10,15,20,25,30,35,40,50,60]
age_labels = ["10-15", "16-20", "21-25", "26-30", "31-35", "36-40", "41-50", "51-60"]

athlete_data_df["age_ranges"] = pd.cut(athlete_data_df["age"],age_bins, labels=age_labels)

athlete_data_df.head(10)
```

```python
# Clean the data by dropping duplicates and setting the index
athlete_data_df.set_index("athlete_id", inplace=True)
noc_data_df.set_index("noc_id", inplace=True)
```

```python
#check the dataframe
athlete_data_df.describe()
```

# Create database connection

```python
# Confirm tables
engine.table_names()
```

Load DataFrames into SQL database

```python
athlete_data_df.to_sql(name='athlete', con=engine, if_exists='append', index=True)
```

```python
noc_data_df.to_sql(name='noc', con=engine, if_exists='append', index=True)
```
